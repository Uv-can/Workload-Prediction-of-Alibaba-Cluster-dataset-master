```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import rcParams
rcParams['figure.figsize']=10,6
import csv
import numpy as np
import math
```

```python
df2=pd.read_csv('final_input_file.csv')
#df2
```

# Mean Based Predictor

```python
#Computational Time
import time
time_start = time.clock()

#Code
X = df2.no_of_instances
predictions=list()
times=list()
SE=0
sumAPE=0
sum=0
for i in range (0,len(X)):
    sum=sum+df2.no_of_instances[i]
meanVal=sum/len(X)
print('Mean=',meanVal)

for i in range (0,len(X)):
    predictions.append(meanVal)
    times.append(df2.time_interval[i])
    print('Time, predicted, expected \n',df2.time_interval[i], meanVal, df2.no_of_instances[i])
    e=abs(meanVal-df2.no_of_instances[i])
    SE=SE+e*e
    APE=abs((meanVal-df2.no_of_instances[i])/df2.no_of_instances[i])
    sumAPE=sumAPE+APE
    #ASE=abs((yhat-obs)/(test[t]-df2.no_of_instances[t-1]))
    #sumMASE=sumMASE+ASE
RMSE=math.sqrt(SE/len(X))
MAPE=(sumAPE/(len(X)))*100
#MASE=(sumMASE/(len(X)))
print('RMSE Value= ', RMSE)
print('MAPE Value= ', MAPE)
#print('MASE Value= ', MASE)
    
#print('Time= \n', times)
#print('Predictions= \n', predictions)

#Computational Time
time_elapsed = (time.clock() - time_start)
print('Computational Time=', time_elapsed)

# plot
fig=plt.figure(figsize=(12,8))
plt.title("Predicted vs Expected- - Mean predictor")
plt.xlabel("Time interval")
plt.ylabel("Number of instances")
plt.plot(times, predictions, color='red', label='Predicted', linestyle='solid')
plt.scatter(df2.time_interval, df2.no_of_instances, color='blue', label='Original')
plt.legend()
plt.show()

```
