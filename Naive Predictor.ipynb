# Naive Predictor

```python
from sklearn.metrics import mean_squared_error

#Computational Time
import time
time_start = time.clock()

#Code
X=df2.no_of_instances
predictions=list()
test=list()
forecast=0
t=0
predictionTime=[]
Begin=1
End=len(X)
SE=0
sumAPE=0
sumMASE=0

for i in range (Begin, End):
    forecast=X[i-1]
    predictions.append(forecast)
    t=df2.time_interval[i]
    predictionTime.append(t)
    test.append(df2.no_of_instances[i])
    print('Time, predicted, expected \n', df2.time_interval[i], forecast, df2.no_of_instances[i])
    #Evaluation
    e=abs(forecast-df2.no_of_instances[i])
    SE=SE+e*e
    APE=abs((forecast-df2.no_of_instances[i])/df2.no_of_instances[i])
    sumAPE=sumAPE+APE
    #ASE=abs((forecast-df2.no_of_instances[i])/(df2.no_of_instances[i]-df2.no_of_instances[i-1]))
    #sumMASE=sumMASE+ASE
#error = math.sqrt(mean_squared_error(test, predictions))
#print('Test MSE: %.3f' % error)
RMSE=math.sqrt(SE/((len(X))-Begin))
MAPE=sumAPE/((len(X))-Begin)*100
#MASE=sumMASE/(len(X)-Begin)
print('RMSE Value= ', RMSE)
print('MAPE Value= ', MAPE)
#print('MASE Value= ', MASE)

#plot graph
fig=plt.figure(figsize=(12,8))
plt.title("Predicted vs Expected- Naive Predictor")
plt.xlabel("Time interval")
plt.ylabel("Number of instances")
plt.plot(predictionTime, predictions, color='red', label='Predicted')
plt.scatter(df2.time_interval, df2.no_of_instances, color='blue', label='Original',linestyle='solid')
plt.legend()
plt.show()

#Computational Time
time_elapsed = (time.clock() - time_start)
print('Computational Time=', time_elapsed)
    
```
